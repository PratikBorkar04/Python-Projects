```python
import tkinter as tk
from tkinter import *
import os
import logging
logging.basicConfig(filename = "f_Reader.log",level = logging.DEBUG , format = "%(asctime)s %(message)s")
base = Tk()
base.geometry("620x520+120+85")
base.title("Reader")
    
l1 = Label(base,text = "Enter Directory :-",font = ("Arial Bold",10))
l1.place(x=10,y=15)

e1 = Entry(base , width = 55)
e1.place(x= 140, y=17)

l2 = Label(base,text = "Enter Extension (Example .txt) :-",font = ("Arial Bold",10))
l2.place(x=10,y=52)

e2 = Entry(base , width = 35)
e2.place(x= 240, y=52)

e3 = Label(base,text = "Save As :-",font = ("Arial Bold",10))
e3.place(x = 50 , y = 450)

l3 = Entry(base , width = 35)
l3.place(x= 150, y=452)

ls = []
l1 = []
    
def method1():
    z= str(e1.get())
    n = str(e2.get())
    try :
        logging.info('method1')
        logging.info('Input from directory')
        a = os.listdir(z)
        for i in a:
            if i.endswith(n):
                l1.append(i)        
    except Exception as ex:
        logging.exception(str(ex))
        print(ex)

    try:
        for j in range(len(l1)):
            a = z+"/"+l1[j]
            f = open(a,"r")
            u = f.read()
            a = (l1[j])
            ls.append(a)
            ls.append(u)
        
    except Exception as ex:
        logging.exception(str(ex))
        print(ex)
    txt1.insert(tk.END,ls)
    
def method2():
    logging.info('method2')
    logging.info('Clear data')
    txt1.delete("1.0","end")
    
def method3():
    s3 = str(l3.get())
    try:
        logging.info('method3')
        logging.info('')
        f = open(s3,"w")
        listToStr = ' '.join(map(str, ls))
        f.write(listToStr)
    except Exception as ex:
        logging.exception(str(ex))
        print(ex)
        
b1 = Button(base, text = "Browse" ,width = "10", background = "green" , foreground = "white",font = ("Arial Bold",10), command = method1)
b1.place(x = 490 , y = 35)

txt1 = Text(base, height = 18 ,width = 80,font = ("Arial Bold",10))
txt1.place(x = 20, y = 80)

b2 = Button(base, text = "Clear" ,width = "10", background = "Yellow" , foreground = "Black",font = ("Arial Bold",10),command = method2)
b2.place(x = 250 , y = 390)

b3 = Button(base, text = "Save" ,width = "10", background = "orange" , foreground = "Black",font = ("Arial Bold",10),command = method3)
b3.place(x = 400 , y = 447)
base.mainloop() 
```
