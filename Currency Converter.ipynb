```python
#imported required modules

from tkinter import *
from tkinter import ttk
from forex_python.converter import CurrencyRates

# main window
base1 =Tk()
base1.geometry("392x400+380+85")
base1.title("Currency Converter")

#Create "Currency Converter" Label
Label1 = Label(base1 , text = "Currency Converter" , font = ("Georgia" , "16"),bg = "#3A3B3C",fg= "white")
Label1.place(x = 102 , y = 15)

#Create "----" Label
Label2 = Label(base1 , text = "-----------------------" , font = ("Georgia" , "16"),bg = "#3A3B3C",fg= "white")
Label2.place(x = 102 , y = 45)

#Create "TEnter Amount" Label
Label3 = Label(base1 , text = "Enter Amount" , font = ("Georgia" , "14"),bg = "#3A3B3C",fg= "white")
Label3.place(x = 130 , y = 72)

#Create "Amount" Entry
Ent1 = Entry(base1,width = 18,bg= "white",font = ("Arial" , "10"))
Ent1.place(x = 130 , y = 114)

#Create "From" Label
Label4 = Label(base1 , text = "From" , font = ("Georgia" , "14"),bg = "#3A3B3C",fg= "white")
Label4.place(x = 170 , y = 138)

#Create dropdownbox 1
a1 = tk.StringVar()
countries1 = ttk.Combobox(base1, width = 18,textvariable = a1)
countries1['values'] = ("USD","INR","GBP","JPY","EUR")
countries1.place(x = 130 , y = 180)
countries1.current(0)

#Create "To" Label
Label5 = Label(base1 , text = "To" , font = ("Georgia" , "14"),bg = "#3A3B3C",fg= "white")
Label5.place(x = 180, y = 205)

#Create dropdownbox 2
a2 = tk.StringVar()
countries2 = ttk.Combobox(base1, width = 18,textvariable = a2)
countries2['values'] = ("USD","INR","GBP","JPY","EUR")
countries2.place(x = 130 , y = 240)
countries2.current(1)

#converting data
def convert():
    c = CurrencyRates()
    x1 = Ent1.get()
    #convert String to float (x1(string) --- > x(float))
    x = float(x1)
    
    #Get data from dropdownbox1
    a1_box = a1.get()
    
    #Get data from dropdownbox2
    a2_box = a2.get()
    
    #Conversion of Amount 
    change = c.get_rate(a1_box, a2_box)
    
    #Multiply Enter amount to Current Converted Amount
    converted_Amount = change * x 
    print(converted_Amount)
         
    #Create "Converted Amount "Label
    label10 = Label(base1,width = 25,bg= "white",font = ("Arial" , "12"),text = converted_Amount , fg = "#3A3B3C")
    label10.place(x = 80 , y = 335)
    
#clearing data
def clear_data():
    Ent1.delete(0,END)
    
#Create Convert Button
Button1 = Button(base1 , width = '10', text = "Convert" , bg = "Green" , fg = "White",command = convert)
Button1.place(x = 80,y = 290)

#Create Clear Button
Button2 = Button(base1 , width = '10', text = "Clear" , bg = "Red" , fg = "White",command = clear_data)
Button2.place(x = 230,y = 290)

#Background color
base1.configure(bg = "#3A3B3C")
base1.mainloop()
```
